```python
def jaccard_similarity(s1, s2):
  # 분모가 0이면 계산할 수 없기 때문에 s1s2 합집합의 크기가 0인 경우 return 0
    if len(s1|s2) == 0:
        return 0
  # 아닌 경우 교집합/합집합 반환
    return len(s1&s2)/len(s1|s2)
    
def pearson_similarity(u1, u2):
    with warnings.catch_warnings():
        warnings.simplefilter("ignore", category=RuntimeWarning)
      # Centered 방식 : 각 벡터에서 NaN이 아닌 값들을 평균낸 값을 빼준다.
        
        u1_mean = np.nanmean(u1)
        u2_mean = np.nanmean(u2)

        if u1_mean == 100 :
            u1_c = u1
        else :
            u1_c = u1 - u1_mean

        if u2_mean == 100 :
            u2_c = u2
        else :
            u2_c = u2 - u2_mean


      # 여기서부턴 Cosine 유사도 구하기
        denom = np.sqrt(np.nansum(u1_c ** 2) * np.nansum(u2_c ** 2))

        if denom != 0:
            return np.nansum(u1_c * u2_c)/denom
        else:
            return 0
```

```python
# corp : 찾으려고 하는 corp
# n : 가장 비슷한 corp를 몇 개 출력할 것인지
# alpha : 유사도 반영 비율

def find_similar_corp (df, corp , alpha):
    
    s1 = set(df.loc[corp,df.loc[corp,:].notnull()].index.values)
    u1 = df.loc[corp,:].values
    
    result = []
    
    col = df.index.values

    for this_corp in col :
        if this_corp[:-2] == corp[:-2] :#같은 회사면 건너뛰기
            continue
        
        s2 = set(df.loc[this_corp,df.loc[this_corp,:].notnull()].index.values)
        u2 = df.loc[this_corp,:].values
        
    
        jaccard = jaccard_similarity(s1, s2) #교집합/합집합
        pearson = pearson_similarity(u1, u2) #벡터 계산
    
    
        score = alpha * pearson + ( 1-alpha) * jaccard
        result.append( (this_corp, round(score,5)))
        
    result.sort(key= lambda r: r[1], reverse= True)
    result_df = pd.DataFrame(result, columns=["corp","score"])
    join_sim_corp_df = pd.merge(result_df.iloc[:100,:], corp_fin[["사업자번호","휴폐업유무"]], how="left", left_on="corp", right_on="사업자번호")
    sum_result=join_sim_corp_df["휴폐업유무"].sum()    

    return join_sim_corp_df, sum_result
```
